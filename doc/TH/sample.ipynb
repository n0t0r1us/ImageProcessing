```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
def image_translate(image,shiftX,shiftY):

    M = np.float32([[1, 0, shiftX], [0, 1, shiftY]])
    return cv2.warpAffine(image,M,(image.shape[1],image.shape[0]))

img=cv2.imread('D:/beach.png')
fig=plt.figure()
ax1,ax2=fig.subplots(1,2)
ax1.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
ax1.set_title("orignal")
# shift image along x,y axis.
pixel_down=20
pixel_right=10
shifted = image_translate(img,pixel_down,pixel_right)
ax2.imshow(cv2.cvtColor(shifted,cv2.COLOR_BGR2RGB))
ax2.set_title("Translated image")
#plt.savefig("translatedi.pdf",bbox_inches=’tight’)
plt.show()
```

```python
after_img=shifted
after_img[0:pixel_down+1, :]=255
after_img[:,0:pixel_right+10,]=255
fig=plt.figure(figsize=(16,9))
ax1,ax2=fig.subplots(1,2)
ax1.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
ax2.imshow(cv2.cvtColor(shifted,cv2.COLOR_BGR2RGB))
plt.show()
```
